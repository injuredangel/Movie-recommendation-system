```python
def trans_float(x):
    return (0 if x=='?' else float(x))
```

```python
def extract_label(field):
    '''提取目标值'''
    label = field[-1]
    return float(label)
```

```python
def extract_feature(filed,cateMap,featureend):
    '''
    filed:每一项数据
    cateMap:字典
    featureend：提取特征字段的结尾
    '''
    #提取分类特征字段
    catex = cateMap[filed[3]]
    catefeature = np.zeros(len(cateMap))
    catefeature[catex]=1
    #提取数值特征
    numfeature = [trans_float(obj) for obj in filed[4:featureend]]
    #返回提取完成的特征
    return np.concatenate((catefeature,numfeature))
```
